```python
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

def graficosOrdenacao(df):
  tipos_ordenacao = ['select_sort', 'insert_sort', 'bubble_sort', 'merge_sort', 'shell_sort', 'quick_sort', 'heap_sort']
  sequencias = ['crescente', 'decrescente', 'aleatorio']
  colunas = ['tipo_ordenacao', 'tamanho_sequencia', 'tempo_execucao', 'tipo_sequencia']

  for tipo_ordenacao in tipos_ordenacao:
    fig, ax = plt.subplots(figsize=(10, 6))
    for sequencia in sequencias:
      # Filtrar os dados usando o Pandas
      dados_filtrados = df[(df[colunas[0]] == tipo_ordenacao) & (df[colunas[3]] == sequencia)]

      # Extrair as colunas do DataFrame filtrado
      tamanho_sequencia = dados_filtrados[colunas[1]]
      tempo_execucao = dados_filtrados[colunas[2]]

      # Plotar o gráfico
      ax.plot(tamanho_sequencia, tempo_execucao, label=sequencia)
      ax.scatter(tamanho_sequencia, tempo_execucao)

    ax.set_xlabel('Tamanho da Sequência')
    ax.set_ylabel('Tempo de Execução')
    ax.set_title(tipo_ordenacao)
    ax.grid(True)
    ax.legend()
    ax.set_yticks(range(1, 30, 1))
    ax.set_xticks(range(10000, 110000, 10000))

  plt.show()


def media_aleatorio(df):
  tipos_ordenacao = ['select_sort', 'insert_sort', 'bubble_sort', 'merge_sort', 'shell_sort', 'quick_sort', 'heap_sort']
  tamanhos_sequencia = df['tamanho_sequencia'].unique()

  medias_tempo_execucao = []

  for tipo_ordenacao in tipos_ordenacao:
    medias = []
    for tamanho_sequencia in tamanhos_sequencia:
      dados_filtrados = df[(df['tipo_sequencia'] == 'aleatorio') & (df['tipo_ordenacao'] == tipo_ordenacao) & (df['tamanho_sequencia'] == tamanho_sequencia)]
      media_tempo_execucao = dados_filtrados['tempo_execucao'].mean()
      medias.append(media_tempo_execucao)
    
    medias_tempo_execucao.append(medias)

  plt.subplots(figsize=(10, 6))
  for i, tipo_ordenacao in enumerate(tipos_ordenacao):
    plt.plot(tamanhos_sequencia, medias_tempo_execucao[i], marker='o', label=tipo_ordenacao)

  plt.xlabel('Tamanho da Sequência')
  plt.ylabel('Tempo de Execução Médio')
  plt.title('Média do Tempo de Execução para Sequências Aleatórias')
  plt.legend()
  plt.grid(True)
  plt.yticks(range(1, 30, 1))
  plt.xticks(range(10000, 110000, 10000))
  plt.show()


if __name__ == '__main__':
  # Ler os dados do arquivo CSV usando o Pandas
  df = pd.read_csv('dados.csv')

  graficosOrdenacao(df)

  media_aleatorio(df)

```
